```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation

import numpy as np
import time

def plot_squares(ax,n,side,colors):
    for i in range(n):
        for j in range(n):
            ax.add_patch(patches.Rectangle((i*side,j*side),side,side,
                                            facecolor=colors[(i+j)%len(colors)]))
    ax.set_xlim([0,n*side])
    ax.set_ylim([0,n*side])

fig,ax = plt.subplots(1)
n = 5
side = 1.0
colors = ['red','blue','green','yellow']
plot_squares(ax,n,side,colors)

def update(frame):
    global n,side
    n += 1
    side = side/2
    ax.clear()
    plot_squares(ax,n,side,colors)

ani = animation.FuncAnimation(fig,update,frames=500,interval=50,blit=False)
plt.show()
```

```python
def grey2rgb(g):
    S = g.shape[0]
    return g[None].expand(3, S, S)
```

```python
def get_uvs(H, W, device='cpu'):
    xs = torch.linspace(-1, 1, W).to(device)
    ys = torch.linspace(-1, 1, H).to(device)

    x, y = torch.meshgrid(xs, ys, indexing='xy')
    uvs = torch.stack([x,y], dim=-1)
    return uvs
```
