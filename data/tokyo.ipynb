```python
# Define the region you want to crop
size = 512
top = int(19456//4*(.63))
left = int(51200//4*(.375))

right = left + size
bottom = top + size

# Crop the image
cropped_image = full_pil.crop((left, top, right, bottom))
cropped_image
```
